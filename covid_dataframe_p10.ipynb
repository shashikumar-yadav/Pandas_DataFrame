P10: Creating DataFrame and performing operations on it

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating dataFrame and printing it
covid_test=pd.DataFrame({
    "Number": [1,2,3,4,5,6],
    "State":["Maharashtra","Karnataka","Gujrat","Delhi","MP","Kerala"],
    "Active_Cases": [2000,1278,694,235,548,1836],
    "Positive":[36,45,100,150,65,87]
})

covid_test
```

```python
#display last 2 record of df

covid_test.tail(2)
```

```python
#sorting DF by number of Positive cases in descending order

covid_test.sort_values(by="Positive", ascending=False)
```

```python
#checking states where positive cases in equal or more than 100


high_cases=covid_test[covid_test["Positive"] >=100]

high_cases
```

```python
#checking statses where postive cases are less than 100 and active cases are more than 1000

check_record=covid_test[(covid_test["Positive"]<100 ) & (covid_test["Active_Cases"]>1000)]

check_record
```

```python
#plotting bar chart by active cases and positive cases in each state

covid_test.plot(kind="bar",x="State",y=["Active_Cases","Positive"])
plt.show()
```
