```python
#1st DataFrame

import pandas as pd

df1 = pd.DataFrame ({
    "Student_ID" : [1,2,3],
    "Name" : ["Alice", "Bob", "Charlie"]
})

print(df1)
```

```python
df2 = pd.DataFrame({
    "Student_ID" : [1,2,3],
    "Grade" : ["B", "A", "A+"]

})

print(df2)
```

```python
merged_dataframe = pd.merge (df1, df2, on = "Student_ID")
print(merged_dataframe)
```
